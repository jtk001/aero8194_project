```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

plt.style.use('bmh')

(x_train,_),_ = tf.keras.datasets.mnist.load_data()
x_train = x_train / 255.
x_hist,x_bins = np.histogram(x_train,bins=30,range=[-0.1,1.1])
widths = np.diff(x_bins)
x_hist = x_hist / np.sum(widths * x_hist)
```

```python
fig,ax = plt.subplots()
ax.bar(x=x_bins[:-1],height=x_hist,width=widths,align='edge',edgecolor='k',linewidth=1,zorder=3)
ax.set_xlabel('Pixel Intensity')
ax.set_ylabel('Normalized Counts')
ax.set_title('MNIST Digits Dataset')
plt.savefig('plots\\mnist_dist.svg')
```

```python
n_binary = np.sum(np.logical_or(x_train == 1., x_train == 0.))
print('{0:.4f}% of pixels are binary'.format(n_binary/len(x_train.ravel())*100))
```
