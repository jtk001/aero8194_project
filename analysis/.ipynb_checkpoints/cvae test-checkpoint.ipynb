```python
import os
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.gridspec import GridSpec
from tensorflow.keras.datasets.mnist import load_data
from tensorflow.keras.models import load_model
from vae import KLDivergenceLayer, ReconstructionLossLayer, SamplingLayer

model_dir = 'models'
model_name = 'conditional_decoder.h5'
model_path = os.path.join(model_dir,model_name)
plot_dir = 'analysis\\plots'
if not os.path.isdir(plot_dir):
    os.makedirs(plot_dir)
```

```python
custom_objects = {'KLDivergenceLayer':KLDivergenceLayer,
                  'ReconstructionLossLayer':ReconstructionLossLayer,
                  'SamplingLayer':SamplingLayer}
decoder = load_model(model_path,custom_objects=custom_objects)
```

```python
decoder.summary()
```

```python
np.random.seed(1)
num = 5
n_classes = 10
latent_dim = 12
z_test = np.random.normal(size=(num*n_classes,latent_dim))
aux_input = np.zeros(shape=(num*n_classes,n_classes))
idx = np.r_[[5*[i] for i in range(10)]]
aux_input[np.arange(num*n_classes),idx.ravel()] = 1
```

```python
preds = decoder.predict([z_test,aux_input]).reshape(-1,28,28)
```

```python
n_images = len(preds)
fig = plt.figure(figsize=(12,30))
gs = GridSpec(n_classes * 3, num)
k = 0
ylims = [z_test.min(),z_test.max()]
for i in range(n_classes):
    for j in range(num):
        ax1 = fig.add_subplot(gs[3*i:3*i+2,j])
        ax1.imshow(preds[k])
        ax1.set_xticks([])
        ax1.set_yticks([])
        ax2 = fig.add_subplot(gs[3*i+2,j])
        ax2.plot(z_test[k])
        ax2.set_xticks([])
        ax2.set_ylim(ylims)
        k+=1
gs.tight_layout(fig)
fname = 'cvae_test.svg'
plt.savefig(os.path.join(plot_dir,fname))
```
