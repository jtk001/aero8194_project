```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('bmh')
plt.rcParams['font.size'] = 12
```

```python
data_dir = os.path.normpath('../train_data')
data_files = [os.path.join(data_dir,i) for i in os.listdir(data_dir) if i.endswith('csv')]
data_file_names = [i.split('.')[0] for i in os.listdir(data_dir) if i.endswith('.csv')]
data = {i:pd.read_csv(j, float_precision='high') for i,j in zip(data_file_names, data_files)}
for name, df in data.items():
    print(name)
    print(df.head())
```

```python
fig_1, ax_1 = plt.subplots(figsize=(6,4))
ax_1.plot(data['vae_train']['epoch'], data['vae_train']['loss'], linewidth=2, label='Train Loss')
ax_1.plot(data['vae_train']['epoch'], data['vae_train']['val_loss'], linewidth=2, label='Test Loss')
ax_1.set_xlabel('Epoch')
ax_1.set_ylabel('$\\overline{\\mathcal{L}}_{\\mathbf{\\theta}}$')
ax_1.set_title('Variational Autoencoder Learning Curves')
ax_1.legend()
```

```python
fig_2, ax_2 = plt.subplots(figsize=(6,4))
ax_2.plot(data['cvae_train']['epoch'], data['cvae_train']['loss'], linewidth=2, label='Train Loss')
ax_2.plot(data['cvae_train']['epoch'], data['cvae_train']['val_loss'], linewidth=2, label='Test Loss')
ax_2.set_xlabel('Epoch')
ax_2.set_ylabel('$\\overline{\\mathcal{L}}_{\\mathbf{\\theta}}$')
ax_2.set_title('Conditional Variational Autoencoder Learning Curves')
ax_2.legend()
```
